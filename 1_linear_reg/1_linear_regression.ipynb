```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
#%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
df.describe()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
#df["price"]
df.price
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df.values,price)
```

**(1) Predecir el precio de una vivienda con área = 3300 pies cuadrados**

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

**Y = m * X + b (m es el coeficiente y b es la intersección)**

```python
3300*135.78767123 + 180616.43835616432
```

**(2) Predecir el precio de una vivienda con área = 5000 pies cuadrados**

```python
reg.predict([[5000]])
```
